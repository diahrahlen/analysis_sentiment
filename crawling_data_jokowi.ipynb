```python
# buat file
csvFile = open('file_jokowi.csv', 'w', encoding='utf-8')

csvwriter = csv.writer(csvFile)

created = []
user_location = []
source = []
id = []
username = []
lang = []
favorite_count = []
retweet_count = []
followers_count = []
statuses_count = []
friends_count = []
text = []

#Use csv writer
csvWriter = csv.writer(csvFile)
for tweet in tweepy.Cursor(api.search, q=search_key, tweet_mode="extended",lang="id",since='2020').items(100):
    created.append(tweet.created_at)
    user_location.append(tweet.user.location)
    source.append(tweet.source)
    id.append(tweet.id)
    username.append(tweet.user.name)
    lang.append(tweet.lang)
    favorite_count.append(tweet.favorite_count)
    retweet_count.append(tweet.retweet_count)
    followers_count.append(tweet.user.followers_count)
    statuses_count.append(tweet.user.statuses_count)
    friends_count.append(tweet.user.friends_count)
    text.append(tweet.full_text)
    
    tweets = [tweet.created_at, tweet.source, tweet.user.location, tweet.id, tweet.user.name, tweet.lang, tweet.favorite_count, tweet.retweet_count, tweet.user.followers_count, tweet.user.statuses_count, tweet.user.friends_count, tweet.full_text.encode("utf-8")]
    csvWriter.writerow(tweets)

csvWriter = csv.writer(csvFile)
csvFile.close()

dictTweets = {"time":created, "source":source, "location":user_location, "id":id, "username":username, "language":lang, "favorite_count":favorite_count,"retweet_count":retweet_count, "user_followers_count":followers_count, "user_statuses_count":statuses_count, "user_friends_count":friends_count, "text":text}
df = pd.DataFrame(dictTweets,columns=["time","location","source","id","username","language","favorite_count","retweet_count","user_followers_count", "user_statuses_count", "user_friends_count","text"])
df
```
