```python
#import library yang dibutuhkan
import matplotlib.pyplot as plt
import squarify # pip install squarify
import pandas as pd
```

```python
df = pd.read_csv('text_count_1.csv')
df.head()
```

```python
#Treemap (menampilkan text yang digunakan sesuai urutan 1-20)
fig, ax = plt.subplots(1, figsize = (12,12))

squarify.plot(sizes=df['count'].head(20), label=df['text_name'], alpha=.8 )
plt.axis('off')
plt.show()
```

```python
#mengurutkan jumlah text terbanyak
df.sort_values('count', ascending=False, inplace=True)
df.head()
```

```python
#Treemap (menampilkan 5 text yang paling banyak digunakan)
fig, ax = plt.subplots(1, figsize = (12,12))

squarify.plot(sizes=df['count'].head(20), label=df['text_name'][:5], alpha=.8 )
plt.axis('off')
plt.show()
```
